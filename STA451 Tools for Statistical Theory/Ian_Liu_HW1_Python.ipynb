1. Please create two upper triangle matrices A and B, calculate C=AB, check C, weather it is an upper triangle matrix, especially the relationship between a_{ii}, b_{ii} and c_{ii}. order n should great than or equal to 6.

2. Please create two unit upper triangle matrices A and B, calculate C=AB, check C, weather it is a unit upper triangle matrix. order n should great than or equal to 6.

```python
import numpy as np
```

```python
ORDER = 9

def create_upper_triangular_matrix(randint_max, order):
  return np.triu(np.random.randint(randint_max, size = (order, order))).copy()

for i in range(5):
  A = create_upper_triangular_matrix(5, ORDER)
  B = create_upper_triangular_matrix(10, ORDER)

  C = A @ B

  print(A)
  print(B)
  print("\nC")
  print(C)
  assert(np.allclose(np.diagonal(C), np.diagonal(A) * np.diagonal(B)) == True)
  assert(np.allclose(C, np.triu(C)) == True) # Returns error if C is not upper triangular matrix

print("All C's were upper triangular matrices!")
```

```python
ORDER = 9

def create_unit_upper_triangular_matrix(randint_max, order):
  result = create_upper_triangular_matrix(randint_max, order)
  np.fill_diagonal(result, 1)
  return result.copy()

for i in range(10):
  A = create_unit_upper_triangular_matrix(5, ORDER)
  B = create_unit_upper_triangular_matrix(10, ORDER)

  C = A @ B

  print(A)
  print(B)
  print("\nC")
  print(C)

  correct_C = np.triu(C.copy())
  np.fill_diagonal(correct_C, 1)

  assert(np.allclose(C, correct_C) == True) # Returns error if C is not unit upper triangular matrix

print("All C's were unit upper triangular matrices!")
```
