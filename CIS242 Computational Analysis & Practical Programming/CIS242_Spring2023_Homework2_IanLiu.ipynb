```
# Part a
import numpy as np
import matplotlib.pyplot as plt

def CIS242_Fixed_Point_Iteration(g, x0: float, n: int = 1000) -> float:
  """Solves a root for a given function using fixed point iteration 

    Parameters
    ----------
    g: function
        Iteration function
    x0: float
        Initial guess of root 
    n: int
        Number of iterations

    Returns
    -------
    xsol: float
        Estimation of root(s) after n interations if the result is not too large ( |xn|â‰¤1000 ). 

    Raises
    ------
    Error
        If divergence occurs (|xn| too large > 1000)
  """

  xnew = x0
  fx = lambda x: x ** 4 - 2 * (x **3) + x
  while n > 0:

    if abs(xnew) > 1000:
      raise Exception("Divergence occured, no root found")

    if fx(xnew) == 0:
      return xnew

    xnew = g(xnew)
    n -= 1
  
  else:
    if abs(fx(xnew)) < 0.0001:
      return xnew
    return "No root found"
  
```

```
# Part a
def CIS242_Bisection(f, x0, n: int) -> float:
  """Solves a root for a given function using bisection

    Parameters
    ----------
    f: function
        A function that must have different signs
    x0: list or tuple of 2 float numbers
        Range used for bisection
    n: int
        A positive integer for how many iterations to go through

    Returns
    -------
    xm: float
        The middle point of two limits of bracket after n interations 

    Raises
    ------
    Error
        If the signs of x0 are different (i.e., x[0] and x[1] has different signs)
  """

  a = x0[0]
  b = x0[1]
  fa = f(a)
  fb = f(b)
  if (fb != 0 and fa/fb >= 0) or (fb == 0 and fa >= 0):
    raise Exception("The inputs are not valid.") 

  while n > 0:
    fa = f(a)
    fb = f(b)
    xm = (a+b)/2

    if f(xm) == 0:
      return xm
    if fa == 0:
      return a
    if fb == 0:
      return b
    
    if fa != 0 and f(xm)/fa >= 0:
      a = xm 
    else:
      b = xm
    
    n -=1
  else:
    if abs(f(xm)) < 0.0001:
      return xm
    return "No root was found"
```

Study the **Secant method** introduced in the lab note by yourself. Then use that method to write a function "CIS242_Secant_Method(f, x0, n)" which returns an approximation of a root for a given function $f(x)$ (a function), two initial guesses given in a list of numbers x0 (a list of two float number, e.g., $[1,2]$) **(noting that $f$ does not have to have different signs for this method)**, with n (an integer) iterations.

**Give an example to prove that you code works.**





```
# Insert your code here
# Part a
def CIS242_Secant_Method(f, x0: list, n: int = 1000) -> float:
    """Solves a root for a given function using Secant method

    Parameters
    ----------
    f: function
        A function for which we solve the root 
    x0: list[float]
        Two initial guesses in float format given in a list 
    n: int, optional
        A positive integer for how many iterations to go through

    Returns
    -------
    xn: float
        The middle point of two limits of bracket after n interations if the result is not too lage ( |xn|â‰¤ 1000 ).
    Raises
    ------
    Error
        If no root was found after n interations 
    
    """

    if len(x0) != 2:
      return 

    xnew = x0[1]
    xold = x0[0]
    while n >= 0:
      
      result = xnew - f(xnew) * (xnew- xold) / (f(xnew) - f(xold))

      if abs(f(result)) < 1.0e-10:
        return result

      if abs(f(xnew)) < 1.0e-10:
        return xnew
      if abs(f(xold)) < 1.0e-10:
        return xold

      xold = xnew
      xnew = result
      
      n -=1 
    else:
      if f(result) == 0:
        return result
      raise Exception(f"No root found within iterations. Last number tried was {result}")
```
