a) Roll three fair dice, let the result be X, Y, Z. Simulate the distrubtion of $X\times(Y+Z)$

b) Find our the **theoretical** probabilities of all the values in (a) and print out your result.

```python
total_rolls = 100000000
threedice_roll = rand.randint(1,7,total_rolls) * ( rand.randint(1,7,total_rolls) + rand.randint(1,7,total_rolls))
gc.collect()

sns.histplot(threedice_roll, bins = np.arange(1.5,72.5,1))
plt.show()
print(f'The frequencies of 2-{len(np.bincount(threedice_roll)[1:])} are ', np.bincount(threedice_roll)[1:])
gc.collect()
```

How many total possible results are there?

```python
two_dice_possiblities = []
all_possiblities = []
for i in range(1,7):
  for j in range(1,7):
    two_dice_possiblities.append(i+j)

for k in range(1,7):
  for possibility in two_dice_possiblities:
    all_possiblities.append(k * possibility)

print("All Possible Results")
print(len(all_possiblities))
print(all_possiblities)
print("All Possible Unique Results")
print(len(set(all_possiblities)))
print(set(all_possiblities))
```

```python
print('The relative frequencies are:')
np.bincount(threedice_roll)[1:]/total_rolls
```

```python
gc.collect()
sim_proportion = np.bincount(threedice_roll)[1:]/total_rolls

True_prob = np.round(len(set(all_possiblities)) * sim_proportion).astype(int)
sum_values = range(2,73)

for index, value in zip(sum_values, True_prob):
  print(f"The theoreical probability of seeing a result of {index} is {value}/{len(set(all_possiblities))}")
gc.collect()
```

**Exercise 2**: 

Given ten students and divide them into three groups. What is the probability for two of them, e.g. James and Tom, to be in different groups? Use simulation to answer the question.

```python
#@title
# 10 students in list 
# randomly pick 1-3 and assign to person to dict
# if James and Tom are different, add 1 to counter 
# do above 10,000 times and divide counter by 10,000
```

```python
ten_students = ['James', 'Tom', 'Kevin', 'Kenny', 'Ian', 'Bamboo', 'Jinglei', 'Nikki', 'Erin', 'Kylie']
total_iterations = 1000000
diff_group_count = 0
for i in range(total_iterations):
  # creates a dictionary with keys and values
  student_dict = dict(zip(ten_students,  rand.randint(1, 4, 10)))
  if student_dict.get('James') != student_dict.get('Tom'):
    diff_group_count += 1

print(f"{diff_group_count}/{total_iterations}= {diff_group_count/ total_iterations}")
```

Given a deck of cards (52 cards, excluding jokers). Draw 13 card randomly from the deck, what is the probability of having seven or more cards in any suit? Use simulation to answer the question.

```python
#@title
# 52 int , sample 13 
# if more than or equal to 7 cards are in 1-13, 14-26, 27-40, 41-52
```

```python
clubs = range(1,14)
spades = range(14,27)
hearts = range(27,40)
diamonds = range(40,53)

print(len(clubs), clubs)
print(len(spades), spades)
print(len(hearts), hearts)
print(len(diamonds), diamonds)
```

```python
happened = 0
iterations = 1000000
for i in range(iterations):
  draw = rand.randint(1, 53, 13)
  clubs_count = 0
  hearts_count = 0
  spades_count = 0
  diamonds_count = 0

  for card in draw:
    if card in clubs:
      clubs_count += 1
    if card in hearts:
      hearts_count += 1
    if card in spades:
      spades_count += 1
    if card in diamonds: 
      diamonds_count += 1

  if clubs_count >= 7 or hearts_count >= 7 or spades_count >= 7 or diamonds_count >= 7:
    happened += 1

print(f"{happened}/{iterations}= {happened/iterations}")
```
